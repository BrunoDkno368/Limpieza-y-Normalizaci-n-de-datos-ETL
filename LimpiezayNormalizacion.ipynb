Verificamos varios errores en los datos de la columna nombre

Normalizacion basicas, elim de espacios adelante, atras y simbolos especiales

```python
df['Nombre_limpio'] = df['nombre']
```

```python
df['Nombre_limpio'] = (df['Nombre_limpio'].
                       astype(str).                           #tipo de datos que sea string
                       str.strip().                           #eliminamos espacios al comienzo y al final
                       str.replace(r"\s+", " ", regex=True))  #eliminamos simbolos especiales
```

```python
df["Nombre_limpio"] = df["Nombre_limpio"].replace({"": np.nan, "nan": np.nan})
```

```python
print(display(df[['Nombre_limpio','nombre']]))
```

Vemos varios errores, 2 registros NaN, Ju@n, Mar ia. vamos a corregir los errores tratando de no perder ninguno ya que son pocos registros

Nan lo reemplazamos por Sin datos

```python
df['Nombre_limpio'].fillna('Sin datos', inplace=True)
```

Ju@n reemplazamos por Juan

primero limpiamos los simbolos que no corresponden

```python
df["Nombre_limpio"] = df["Nombre_limpio"].str.replace(
    r"[^A-Za-zÁÉÍÓÚáéíóúÑñ ]",
    "",
    regex=True
)
```

paso de Ju@n a Jun aun tenemos que reeplazar para que quede unificado

```python
df['Nombre_limpio'] = df['Nombre_limpio'].replace({'Jun':'Juan'})
```

queda totalemente limpio Ju@n en Juan

pasamos de Mar ia a María

```python
df['Nombre_limpio'] = df['Nombre_limpio'].replace({'Mar ia':'María'})
```

queda remplazado el dato Mar ia

##Columna MONTO DE COMPRA

Verificamos el typo de la columa

```python
df['monto_compra'].dtypes
```

```python
df['monto_compra'].unique()
```

```python
df['monto_limpio'] = (df['monto_compra']
                      .astype(str)                  #Pasamos a string los datos de la columna seleccionada
                      .str.strip())                   #Eliminamos los datos al comienzo y al final
```

pasamos a String ya que como los dato estan sucios no estan en el mismo formato lo unificamos para poder trabajarlo mejor

limpiamos las columnas vacias y las columas nan en NaN

```python
print(display(df[['monto_limpio', 'monto_compra']]))
```

Tenemos un numero que es 1.200,30 tenemos que logrrar pasarlo 1200.30 si no cuando lo pasemos a numerico va a dar NaN

```python
df["monto_limpio"] = (
    df["monto_limpio"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
)

df["monto_limpio"] = pd.to_numeric(df["monto_limpio"], errors="coerce")
```

Como ya tenemos todos los dato en numericos o Nan vamos a pasarlo a numerico

```python
df['monto_limpio'] = pd.to_numeric(df['monto_limpio'], errors= 'coerce')
```

Vemos que hay un monto que es -300 y un monto 9999999 ya que vemos un saldo negativo y uno muy grande vamos a pasarlo a NaN ya que por regla de negocio no podemos tener una compra negativa y no lopodemos poner como positivo porque puede modificar la calidad de los datos

```python
df.loc[df["monto_limpio"] < 0, "monto_limpio"] = np.nan
df.loc[df["monto_limpio"] > 90000, "monto_limpio"] = np.nan
```

```python
print(display(df[['monto_limpio', 'monto_compra']]))
```

como tenemos varios Nan vamos a redondear con los datos que nos quedan

```python
df['monto_limpio'].fillna(df['monto_limpio'].mean().round(0), inplace=True)
```

con esto concluimos la limpieza de la columna monto de compra

```python
print(display(df[['monto_limpio', 'monto_compra']]))
```

asignamos monto limpio a la columna original de monto de compra

```python
df['monto_compra']=df['monto_limpio']
```

```python
df.drop(['monto_limpio'], axis=1, inplace=True)
```

##Columna FECHA DE REGISTRO

```python
df['fecha_registro'].unique()
```

Vemos varios errores y diferentes formatos 2024-01-05, enero 5 2024, 2024-13-01, 2024-01-05 12:00

Normalizamos todo a texto

```python
df['fecha_str'] = df['fecha_registro'].astype(str)
```

```python

meses = {
    'enero': 'january',
    'febrero': 'february',
    'marzo': 'march',
    'abril': 'april',
    'mayo': 'may',
    'junio': 'june',
    'julio': 'july',
    'agosto': 'august',
    'septiembre': 'september',
    'octubre': 'october',
    'noviembre': 'november',
    'diciembre': 'december'
}

for es, en in meses.items():
    df['fecha_str'] = df['fecha_str'].str.replace(es, en, regex=False)
```

```python
df['fecha_str']
```

```python
df['fecha_str'] = df['fecha_str'].str.replace('/','-',regex=False)
```

```python
df['fecha_str']
```

```python
df['fecha_str'] = df['fecha_str'].replace({'2024-01-05 12:00': '2024-01-05',
                                           '05-01-2024': '2024-01-05',
                                          'january 5 2024':'2024-01-05',
                                          '2024-13-01':'2024-01-13',
                                          '2024-1-5':'2024-01-05'})
```

```python
df['fecha_str']
```

```python
df['fecha_str'] = pd.to_datetime(df['fecha_str'], format='%Y-%m-%d', errors='coerce')
```
